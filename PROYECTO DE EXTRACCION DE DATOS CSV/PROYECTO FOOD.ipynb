## 2.-  Distancia Euclidiana

__La distancia euclidiana__ es la generalizaci贸n del __`teorema de Pit谩goras`__.

$$d_{E}(P,Q)=\sqrt{(p_{1}-q_{1})^{2}+(p_{2}-q_{2})^{2}+...+(p_{n}-q_{n})^{2}}=\sqrt{\sum_{i=1}^{n}}(p_{i}-q_{i})^{2})$$

## 3.- Correlaci贸n de Pearson  (Similitud)

El __coeficiente de correlaci贸n de Pearson__ es una medida de __`dependencia lineal entre dos variables aleatorias cuantitativas`__. 

El coeficiente de correlaci贸n puede tomar un __rango de valores de +1 a -1__. Un valor de 0 indica que no hay asociaci贸n entre las dos variables. __`Un valor mayor que 0 indica una asociaci贸n positiva`__ y __`Un valor menor que 0 indica una asociaci贸n negativa`__.

$$s = \frac{1}{1+d_{E}}$$ 

![linea 2](https://user-images.githubusercontent.com/19308295/115926262-2fb62980-a448-11eb-8189-c2f10e499944.png)

```python
from scipy.spatial.distance import euclidean
```

```python
n = data[data.columns[1:]].to_numpy()
m = data[data.columns[0]].to_numpy()
print(n)
print(m)
```

![linea 2](https://user-images.githubusercontent.com/19308295/115926262-2fb62980-a448-11eb-8189-c2f10e499944.png)

##  Hallando la distancia y similitud  de los encuestados.

```python
from pandas import DataFrame
vector=len(m)
matriz= np.zeros((vector,vector))

for i in range(vector):
    for j in range(vector):
        if j and i== np.NaN:
            continue
        else:
            dist=np.sqrt(np.nansum((n[i]-n[j])**2))
            simil=(1/(1+dist))
            matriz[i][j] = simil

da = pd.DataFrame(matriz,columns=m,index=m)
da
```

```python
cl=da.unstack()
cl.sort_values(ascending=False)[range(29,36)]
```

![linea 2](https://user-images.githubusercontent.com/19308295/115926262-2fb62980-a448-11eb-8189-c2f10e499944.png)

###  Validaci贸n - Matrix de Correlaci贸n

Se realiza la validaci贸n de los resultados obtenidos con la   `Matriz de Correlaci贸n` en `Numpy` 

```python
nampa_matriz=np.corrcoef(da)
print(nampa_matriz)
```

```python
correlacion=pd.DataFrame(nampa_matriz,columns=m,index=m)
correlacion.round(decimals=4)
```

```python
cl=correlacion.abs().unstack()
cl.sort_values(ascending=False)[range(29,36)]
```
