```python
import numpy as np
import pandas as pd
from pandas import DataFrame
```

```python
#archivo csv separado por comas
data = pd.read_csv('hipotetico - Respuestas de formulario 1.csv')

#leer lineas
data
```

```python
from scipy.spatial.distance import euclidean
```

```python
data.notnull().tail()
```

```python
data
```

```python
x=data.fillna(1)

u = x[x.columns[1:]].to_numpy()
v = x[x.columns[0]].to_numpy()
```

```python

from pandas import DataFrame
vector=len(v)
matriz= np.zeros((vector,vector))

for i in range(vector):
    for j in range(vector):
        dist=euclidean(u[i],u[j])
        simil=(1/(1+dist))
        matriz[i][j] = simil
        
da = pd.DataFrame(matriz,columns=v,index=v)
da.round(decimals=4)
```

```python
z, c= u.shape
for p in range(0,z):
    
    for q in range(0,z):
    
        print('relacion entre',v[p],'y',v[q],'\n')
        
        distancia= np.sqrt(np.nansum((u[p]-u[q])**2))            # Distancia euclidiana
        
        print('\t\tLa Distancia es ', distancia)
        
        similitud=1/(1+distancia)                                # Similitud de Pearson
        
        print('\t\tla Similitud es ', similitud)
```
